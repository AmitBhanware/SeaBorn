```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
plt.figure(figsize=(15,9))
sns.scatterplot(x='salary',y='sales', data=df, style='level of education', s=200,hue='level of education')
plt.savefig('first_seaborn.jpg')
```

```python
plt.figure(figsize=(8,5),dpi=100)
sns.rugplot(x='salary', data=df, height=0.5)
```

```python

sns.displot(data=df,x='salary', bins=10, color='red',edgecolor='blue', linewidth=4)
```

```python
sns.displot(data=df, x='salary', color='blue', edgecolor='yellow',kde=True, rug=True)
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
df=pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='math score', y='parental level of education',hue='test preparation course', palette='Set2')

plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data= df, x='math score',y='gender', size=2)
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
sns.jointplot(data=df, x='math score', y='reading score',hue='gender', kind='kde', alpha=0.2, shade= True)
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(data=df, hue='gender', diag_kind='hist', corner=True)
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='violin', row='lunch', hue='lunch',col='lunch')
```

```python
g= sns.PairGrid(df)

g= g.map_upper(sns.scatterplot)

g= g.map_diag(sns.histplot)

g= g.map_lower(sns.kdeplot)
```
